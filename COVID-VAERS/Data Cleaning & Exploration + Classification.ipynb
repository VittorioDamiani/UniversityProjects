## Supervised Learning

```python
#We create a subset with only numerical and categorical variables (drop ID and text columns)
data_sup = clean_data.drop(['VAERS_ID', 'SYMPTOM_TEXT', 'LAB_DATA', 'OTHER_MEDS', 
                            'CUR_ILL', 'HISTORY', 'PRIOR_VAX', 'SPLTTYPE', 'ALLERGIES'], axis=1) 
data_sup
#data_sup.to_csv('data_sup.csv')
```

```python
#Move 'Serious' (our target) to the end
Serious = data_sup.loc[:,'Serious']
data_sup = data_sup.drop(['Serious'], axis=1)
```

```python
data_sup = pd.concat([data_sup, Serious], axis=1)
data_sup
```

### Scaling

```python
#MinMax scaler
from sklearn.preprocessing import MinMaxScaler

mm_scaler =  MinMaxScaler(copy=False,feature_range=(0, 1))

scaled_data = pd.DataFrame(mm_scaler.fit_transform(data_sup), columns = data_sup.columns)
```

```python
scaled_data
```

```python
print(scaled_data.shape)
```

### Split into Train and Test

```python
#Separate X and y (explanatory variables and target variable)
X = scaled_data.iloc[:,0:-1]
y = scaled_data.iloc[:,-1] 

#X.head()
#y.head()
```

```python
#Visualize the counts for each Target class
target_dist = data_sup.groupby('Serious').size()
print(target_dist)

%matplotlib inline

target_dist.plot.bar(x = '',y = '',title = 'Distribution of target')
```

```python
from sklearn.model_selection import train_test_split

#Split data into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.30,
                                                    stratify = y, 
                                                    random_state = 123)
print(X_train.shape,X_test.shape)
```

### Multi-layer Perceptron Classifier

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(5, 2),
                      activation='relu', alpha=0.01, batch_size='auto',              
                      learning_rate='constant', learning_rate_init=0.001,
                      max_iter=200, solver='lbfgs', tol=0.01,
                      validation_fraction=0.2, verbose=True)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

#### GridSearch

```python
#Define Classifier and Parameters
from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier()
parameters = {"hidden_layer_sizes":[(5, 2),(10,5,2)],  
              "max_iter": [5000], 
              "alpha": [0.1,1,10]}
```

```python
#Define the GridSearch
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(classifier, parameters, cv=3, scoring = 'f1', verbose=50, n_jobs=-1, refit=True)
```

```python
#Fit the GridSearch
gs = gs.fit(X_train, y_train)
```

```python
#Print the results of the GridSearch
print('***GRIDSEARCH RESULTS***')

print("Best score: %f using %s" % (gs.best_score_, gs.best_params_))
means = gs.cv_results_['mean_test_score']
stds = gs.cv_results_['std_test_score']
params = gs.cv_results_['params']

for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
import matplotlib.pyplot as plt

plt.plot (means, color='blue', alpha=1.00)
plt.show()

print(gs.best_score_)
```

#### Test the best model

```python
best_model = gs.best_estimator_
y_pred = best_model.predict(X_test)

y_pred_train = best_model.predict(X_train)
```

#### Metrics

```python
#F1 scores
from sklearn import metrics

print('***RESULTS ON TRAIN SET***')
print("f1_score: ", metrics.f1_score(y_train, y_pred_train))
print("--")
print('***RESULTS ON TEST SET***')
print("f1_score: ", metrics.f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Reds", cbar=False)
```

#### ROC curve and AUC

```python
from sklearn import metrics

model = best_model
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred_train = model.predict(X_train)

y_probs = model.predict_proba(X_test)  

fpr, tpr, thresholds=metrics.roc_curve(y_test,  y_probs[:,1])

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, label='ROC')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

auc = metrics.roc_auc_score(y_test, y_probs[:,1])
print('AUC: %.2f' % auc)
```
